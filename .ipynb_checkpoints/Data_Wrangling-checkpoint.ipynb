```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"

headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration",
           "num-of-doors", "body-style", "drive-wheels", "engine-location", 
           "wheel-base", "length", "width", "height", "curb-weight", "engine-type",
           "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke",
           "compression-ratio", "horsepower", "peak-rpm", "city-mpg",
           "highway-mpg", "price"]

# Use 'names' instead of 'header' to assign headers to the file 
df = pd.read_csv(filepath, names=headers)
df.head()
```

```python
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#Evaluate Missing Data
missing_data = df.isnull()
missing_data.head(5)
```

```python
#Count missing values in each column
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
#Calculate the mean value for the "normalized-losses" column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#Replace "NaN" with the mean value in "normalized-losses" column
df.replace({"normalized-losses": {np.nan: avg_norm_loss}}, inplace=True)
```

```python
#Calculate the mean value for the "bore" column
mean_bore = df['bore'].astype("float").mean(axis=0)
print("Mean of bore:", mean_bore)
```

```python
#Replace "NaN" with the mean value in the "bore" column
df.replace({'bore': {np.nan: mean_bore}}, inplace = True)
```

```python
#See which values are present in a particular column
df['num-of-doors'].value_counts()
```

```python
#Calculate the most common type
most_common_door_type = df['num-of-doors'].value_counts().idxmax()
print(most_common_door_type)
```

```python
#replace the missing 'num-of-doors' values by the most frequent 
df.replace({'num-of-doors': {np.nan: most_common_door_type }}, inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=['price'], axis=0,inplace=True)
# reset index, because we droped two rows
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df.dtypes
```

```python
#Convert data types to proper format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
df.dtypes
```

```python
#Data Standardization
#Transform mpg to L/100km in the column of "highway-mpg" and change the name of column to "highway-L/100km". 
df['highway-mpg'] = 235/df["highway-mpg"]
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace = True)
df.head()
```

```python
#Data Normalization
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height']=df['height']/df['height'].max()
df[['length','width','height']].head()
```

```python
#Binning
#Calculate the mean value for the "horsepower" column
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print('Avg horsepower: ', avg_horsepower)

#Replace "NaN" with the mean value in the "horsepower" column
df.replace({'horsepower': {np.nan: avg_horsepower}},inplace=True)

#Convert data to correct format:
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
#Plot the histogram of horsepower to see the distribution of horsepower.
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Build a bin array with a minimum value to a maximum value. The values will determine when one bin ends and another begins.
bins=np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins
```

```python
#set group names
group_names = ['Low', 'Medium', 'High']
```

```python
#Apply the function "cut" to determine what each value of df['horsepower'] belongs to.
df['horsepower-binned'] = pd.cut(df['horsepower'],bins, labels=group_names, include_lowest = True)
df[['horsepower','horsepower-binned']].head(20)
```

```python
#See the number of vehicles in each bin:
df[['horsepower-binned']].value_counts()
```

```python
#plot the distribution of each bin
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Indicator/Dummy Variables
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel_type_gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```
